```python
#Most common type among Legendary Pokemon
legendary_df.head()
```

```python
all_types = pd.concat([
    legendary_df[['Type 1']].rename(columns={'Type 1': 'Type'}),
    legendary_df[['Type 2']].rename(columns={'Type 2': 'Type'})
]) #Combine into a single column
```

```python
all_types.head()
```

```python
all_types.groupby('Type')['Type'].value_counts().sort_values(ascending = False)
```

```python
#Most power Water from first 3 generations
df.loc[((df['Type 1'] == 'Water') | (df['Type 2'] == 'Water')) & (df['Generation'].isin([1, 2, 3]))].sort_values(by = 'Total', ascending = False).head()
```

```python
#Distribution of pokemon from generations
df['Generation'].value_counts(sort = False).plot(kind = 'bar', color = 'mediumturquoise', edgecolor = 'seagreen')
```

```python
#Most powerful dragon from last 2 Generations
max_gen = df['Generation'].max()
df.loc[((df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon')) & (df['Generation'].isin([max_gen, max_gen - 1]))].sort_values(by = 'Total', ascending = False).head()
```

```python
#Fire Primary Type Pokemon with more than 100 Attack
df.loc[(df['Type 1'] == 'Fire') & (df['Attack'] >= 100)].sort_values(by = 'Attack', ascending = False).head()
```

```python
#All Water-Flying Types
df.loc[(df['Type 1'] == 'Water') & (df['Type 2'] == 'Flying')].head()
```

```python
#Legendary Pokemon of primary type Fire. Select Name, Attack, and Generation Only
df.loc[
    ((df['Type 1'] == 'Fire') & (df['Legendary'])), 
    ['Name', 'Attack', 'Generation']
]
```

```python
#Select the bottom 5% speed and top 5% speed
ax = df['Speed'].plot(kind = 'hist', figsize = (10, 5), bins = 100)
ax.axvline(df['Speed'].quantile(0.05), color = 'red')
ax.axvline(df['Speed'].quantile(0.95), color = 'red')
```

```python
bottom_5 = df['Speed'].quantile(0.05)
top_5 = df['Speed'].quantile(0.95)
print(bottom_5, top_5)
```

```python
df.loc[(df['Attack'] < bottom_5) | (df['Attack'] > top_5)].sample(10)
```

```python
#Most Powerful Legendary Pokemon
sns.scatterplot(data = df, x = 'Defense', y = 'Attack', hue = 'Legendary',
                palette = {True: 'red', False: 'blue'})
```

```python
df['Atk_Def_Sum'] = df['Attack'] + df['Defense']
df.loc[df['Legendary']].sort_values(by = 'Atk_Def_Sum', ascending = False)
```
