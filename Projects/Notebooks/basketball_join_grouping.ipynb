### Analysis

```python
#Team with most players
df.groupby('Team').agg(total_players_per_team = ('name', 'count')).sort_values(by = 'total_players_per_team', ascending = False).head()
```

```python
df.columns
```

```python
#Team with lowest points scored
df.groupby('Team').agg(total_points = ('PTS', 'sum')).sort_values(by = 'total_points', ascending = True).head()
```

```python
#Team with most 3PM
df.groupby('Team').agg(three_pointers = ('3PM', 'sum')).sort_values(by = 'three_pointers', ascending = False).head()
```

```python
#Team with highest FG% = FGM/FGA
fg_per_team = df.groupby('Team').agg({'FGM': 'sum', 'FGA': 'sum'})
fg_per_team.head()
```

```python
fg_per_team['FG%'] = fg_per_team['FGM'] / fg_per_team['FGA']
fg_per_team.sort_values(by = 'FG%', ascending = False).head()
```

```python
#Difference between best and worst 3P shooters by position
pos_3p_accuracy = df.groupby('POS').agg({'3PM': 'sum', '3PA': 'sum'})
pos_3p_accuracy
```

```python
pos_3p_accuracy['3P%'] = pos_3p_accuracy['3PM'] / pos_3p_accuracy['3PA']
pos_3p_accuracy.sort_values(by = '3P%', ascending = False)
```

```python
pos_3p_accuracy['3P%'].max() - pos_3p_accuracy['3P%'].min()
```

```python
#Find the best scorers by points in each team
best_scorer_by_team = df.loc[df.groupby('Team')['PTS'].idxmax(), ['Team', 'name', 'PTS']]
best_scorer_by_team.sort_values(by = 'PTS', ascending = False).head(10)
```

```python
#Which team has the youngest squad by avg age
df.groupby('Team').agg(average_age = ('Age', 'mean')).sort_values(by = 'average_age', ascending = True).head(10)
```
