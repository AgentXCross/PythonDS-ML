MatPlotLib Python Library for Data Science

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 1000) #1000 points from 0 to 10
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x[::10], y[::10], color = 'red')
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.plot(x, np.cos(x), color = 'green')
```
