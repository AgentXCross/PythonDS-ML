```python
m = np.arange(100)
```

```python
n = m[3:9] #Index 3 to 9 not including 9. n is linked to m changing one changes the other
n[0] = -1200
print(n)
```

```python
n = m[3:9].copy() #Used to make unlinked copy
```

```python
o = m[::-1] #-1 to start (reverse)
print(o)
```

```python
p = m[0::2] #Go by every 2
print(p)
```

```python
indices = np.where(m == -1200) #Locate an element by value
print(indices)
```

```python
y = np.random.rand(1000000)
%timeit sum(y)
%timeit np.sum(y)
```

```python
def my_sum(array):
    sum = 0
    for value in array:
        sum += value
    return sum
```

```python
%timeit my_sum(y)
```
